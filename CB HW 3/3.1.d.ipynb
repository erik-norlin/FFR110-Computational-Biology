```python
import numpy as np
import matplotlib.pyplot as plt
# import PyQt5
# import matplotlib as mpl
# # import latex 
# mpl.use('Qt5Agg')


alpha = 2
beta = 1
no_runs = 2000
Tmax = 4000
dt = 0.1
n = np.zeros((no_runs, int(Tmax/dt)))
N = 30
n0 = N*(1-beta/alpha)
n[:,0] = n0
print("N*", N*(1-beta/alpha))

for i in range(no_runs):

    index_old = 0
    t = 0
    if i % 10 == 0: print(i, no_runs)

    while t < Tmax-1:

        if n[i,index_old] < 1:
            break
        else:
            bn = alpha*n[i,index_old]*(1-n[i,index_old]/N)
            dn = beta*n[i,index_old]
            tb = np.random.exponential(1/bn)
            td = np.random.exponential(1/dn)

            t = t + np.minimum(tb, td)
            index = int(np.floor(t/dt) + 1)
            n[i,(index_old + 1):(index - 1)] = n[i,index_old]

            if tb < td:
                n[i,index] = n[i,index_old] + 1
            elif td < tb:
                n[i,index] = n[i,index_old] - 1

            index_old = index

Pn = np.zeros_like(n[0,:])
for i in range(len(Pn)):
    Pn[i] = (np.sum(n[:,i]) / no_runs)

```
